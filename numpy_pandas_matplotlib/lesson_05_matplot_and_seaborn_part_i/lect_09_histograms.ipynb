```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
pokemon_df = pd.read_csv('./pokemon.csv')
# Note :: Uncomment below statement to observe output
#pokemon_df.head(2)
bins = np.arange(0, pokemon_df['speed'].max() + 5, 5)
plt.hist(data = pokemon_df, x = 'speed', bins = bins)
```

```python
# Another way to plot histogram is using Seaborn.displot
sb.distplot(pokemon_df['speed'])
# You may avoid curve using
# sb.distplot(pokemon_df['speed'], kde = False)
```
